```python
import random
import copy
```

```python
rows = int(input("How many rows in the maze? "))
columns = int((input("How many columns in the maze? ")))
```

```python
def makeMaze (rows, columns):
    maze = {}
    maze[rows + 1] = rowWalls (columns)
    for row in range(rows, 0, -1):
        layout = []
        for column in range(1, columns+1):
            specificSpacesCharacteristic = input("What is the space at " + str(column) + ", " + str(row))
            layout.append(specificSpacesCharacteristic)
        maze[row] = layout
    maze[0] = rowWalls (columns)
    for row in maze:
        layout = []
        layout.append("#")
        for value in maze[row]:
            layout.append(value)
        layout.append("#")
        maze[row] = layout
    return maze
```

```python
def rowWalls (columns):
    layout = []
    for column in range(columns):
        layout.append("#")
    return layout
```

```python
maze = makeMaze (rows, columns)
```

```python
def makeQTable (rows, columns):
    action = 4
    qTable = {}
    for row in range(0, rows+2):
        for column in range(0, columns+2):
            actionList = []
            for a in range(action):
                actionList.append(0)
            qTable[column, row] = actionList
    return qTable
```

```python
qTable = makeQTable (rows, columns)
```

```python
def startGame (maze):
    state = 0
    for row in maze:
        for column in range(len(maze[row])):
            if maze[row][column] == 'S':
                state = (column, row)
                maze[row][column] = '@'
    return state, maze
```

```python
def makeActions(qTable, state):
    best = max(qTable[state])
    if not (qTable[state].count(best) == 1):
        indexList = []
        for i in range(0, len(qTable[state])):
            if qTable[state][i] == best:
                indexList.append(i)
        index = random.choice(indexList)
    else:
        index = qTable[state].index(best)
    if index == 0:
        return [0, 1]
    elif index == 1:
        return [0, -1]
    elif index == 2:
        return [-1, 0]
    else:
        return [1, 0]
```

```python
def makeAction (qTable, state):
    from random import randrange
    value = randrange(100)
    if value == 0:
        actions = randomActions()
    else:
        actions = qTableActions(qTable, state)
    return actions
```

```python
def qTableActions(qTable, state):
    best = max(qTable[state])
    index = qTable[state].index(best)
    if index == 0:
        return [0, 1]
    elif index == 1:
        return [0, -1]
    elif index == 2:
        return [-1, 0]
    else:
        return [1, 0]
```

```python
def randomActions():
    actions = [0, 0]
    random_bits = random.getrandbits(1)
    horizontalNotVertical = bool(random_bits)
    vector = random.getrandbits(1)
    if horizontalNotVertical:
        actions[0] = 2*vector-1
    else:
        actions[1] = 2*vector-1
    return actions
```

```python
def applyAction (maze, qTable, actions, state, notWon, savedMaze):
    previousState = state
    state = list(state)
    for xY in range(len(actions)):
        state[xY] = state[xY] + actions[xY]
    state, notWon, qTable = checkIfViable (maze, qTable, state, previousState, savedMaze, actions)
    return tuple(state), notWon, qTable
```

```python
def checkIfViable (maze, qTable, state, previousState, savedMaze, actions):
    if maze[state[1]][state[0]] == '#':
        bellmanEquation (qTable, state, previousState, actions, -1000)
        return previousState, True, qTable
    elif maze[state[1]][state[0]] == 'E':
        bellmanEquation (qTable, state, previousState, actions, 500)
        maze[state[1]][state[0]] = '@'
        if savedMaze[previousState[1]][previousState[0]] == 'S':
            maze[previousState[1]][previousState[0]] = 'S'
        else:
            maze[previousState[1]][previousState[0]] = ' '
        return state, False, qTable
    else:
        bellmanEquation (qTable, state, previousState, actions, -10)
        maze[state[1]][state[0]] = '@'
        if savedMaze[previousState[1]][previousState[0]] == 'S':
            maze[previousState[1]][previousState[0]] = 'S'
        else:
            maze[previousState[1]][previousState[0]] = ' '
        return state, True, qTable
```

```python
def bellmanEquation (qTable, state, previousState, actions, points):
    if actions == [1, 0]:
        location = 3
    elif actions == [-1, 0]:
        location = 2
    elif actions == [0, 1]:
        location = 0
    else:
        location = 1
    maximum = max(qTable[tuple(state)])
    q = (1-0.5)*(qTable[tuple(previousState)][location]) + 0.5 * (points + 0.5*maximum)
    qTable[tuple(previousState)][location] = q
    return qTable
```

```python
def qLearning (maze, qTable):
    savedMaze = copy.deepcopy(maze)
    history = []
    for episodes in range(1, 2000+1):
        maze = copy.deepcopy(savedMaze)
        state, maze = startGame (maze)
        notWon = True
        #for row in maze:
        #    print(maze[row])
        #print("-------------------")
        while notWon:
            actions = makeActions (qTable, state)
            state, notWon, qTable = applyAction (maze, qTable, actions, state, notWon, savedMaze)
            #for row in maze:
            #    print(maze[row])
            #print("-------------------")
            #print(len(history))
        history.append(qTable)
    return qTable, history
```
